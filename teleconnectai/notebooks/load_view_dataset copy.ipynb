```python
def check_numeric_ranges(df):
    checks = {
        "age": (0, 120),
        "tenure_months": (0, 120),
        "monthly_charges": (0, None),
        "total_charges": (0, None),
        "avg_monthly_gb_used": (0, None),
        "avg_monthly_minutes": (0, None),
        "num_support_tickets": (0, None),
        "num_additional_services": (0, None)
    }

    for col, (min_val, max_val) in checks.items():
        invalid = df[col].isna()

        if min_val is not None:
            invalid |= df[col] < min_val

        if max_val is not None:
            invalid |= df[col] > max_val

        print(f"{col:<30} Invalid Rows : {invalid.sum()}")


check_numeric_ranges(df)
```

## Data Cleaning

```python
df_clean = df.copy()
```

```python
df_clean.replace(r'^\s*$', np.nan, regex=True, inplace=True)
```

```python
phone_mapping = {
    "Y": "Yes",
    "YES": "Yes",
    "yes": "Yes",
    "y": "Yes",

    "N": "No",
    "NO": "No",
    "no": "No",
    "n": "No"
}

df_clean["phone_service"] = (
    df_clean["phone_service"]
        .astype(str)
        .str.strip()
        .replace(phone_mapping)
)
```

```python
df_clean["phone_service"].value_counts(dropna=False)
```

```python
df_clean["age"].dropna().head(20)
df_clean["age"] = df_clean["age"].round().astype("Int64")
```

```python
print(df_clean["satisfaction_score"].min())
print(df_clean["satisfaction_score"].max())

df_clean["satisfaction_score"].value_counts().sort_index()
```

```python
df_clean.loc[
    ~df_clean["satisfaction_score"].between(1, 5),
    "satisfaction_score"
] = np.nan
```

```python
df_clean["last_interaction_date"] = pd.to_datetime(
    df_clean["last_interaction_date"],
    errors="coerce"
)
```

```python
quality_summary = pd.DataFrame([
    {
        "Column": "age",
        "Issue": "Missing values",
        "Rows Affected": df_clean["age"].isna().sum(),
        "Cleaning Strategy": "Median Imputation",
        "Reason": "Age distribution is often skewed and median is robust to outliers."
    },
    {
        "Column": "phone_service",
        "Issue": "Inconsistent categorical values",
        "Rows Affected": "Multiple",
        "Cleaning Strategy": "Standardized to Yes/No",
        "Reason": "Ensures consistent encoding before modeling."
    },
    {
        "Column": "satisfaction_score",
        "Issue": "Invalid values outside business range",
        "Rows Affected": (
            (~df_clean["satisfaction_score"].between(1,5)).sum()
        ),
        "Cleaning Strategy": "Replace invalid values with NaN, then impute median",
        "Reason": "Invalid scores likely represent data entry errors or sentinel values."
    }
])

quality_summary
```

```python
numeric_cols = df_clean.select_dtypes(include=["int64", "float64"]).columns

for col in numeric_cols:
    if df_clean[col].isnull().sum() > 0:
        df_clean[col].fillna(df_clean[col].median(), inplace=True)
```

```python
categorical_cols = df_clean.select_dtypes(include="object").columns

for col in categorical_cols:
    if df_clean[col].isnull().sum() > 0:
        df_clean[col].fillna(df_clean[col].mode()[0], inplace=True)
```
